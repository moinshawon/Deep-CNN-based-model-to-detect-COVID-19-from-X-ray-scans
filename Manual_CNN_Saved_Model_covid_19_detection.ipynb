```
# Train from scratch
train_datagen = image.ImageDataGenerator(
    rescale = 1./255,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
)

test_datagen = image.ImageDataGenerator(rescale=1./255)
```

```
batch_size = 32

train_generator = train_datagen.flow_from_directory(
    '/content/drive/MyDrive/Covid19 Detection/Covid Dataset 8/Gray images/For Save/Train',
    target_size = (224,224),
    batch_size = batch_size,
    class_mode = 'binary')

validation_generator = test_datagen.flow_from_directory(
    '/content/drive/MyDrive/Covid19 Detection/Covid Dataset 8/Gray images/For Save/Validation',
    target_size = (224,224),
    batch_size = batch_size,
    class_mode = 'binary')
```

```
# CNN Based Model in Keras

model = Sequential()
model.add(Conv2D(32,kernel_size=(3,3),activation='relu',input_shape=(224,224,3)))
model.add(Conv2D(64,(3,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))

model.add(Conv2D(64,(3,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))

model.add(Conv2D(128,(3,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(64,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1,activation='sigmoid'))

model.compile(loss=keras.losses.binary_crossentropy,optimizer='adam',metrics=['accuracy'])
```

```
leaky_relu_alpha = 0.1

model7 = Sequential()

model7.add(Conv2D(32, kernel_size=(3, 3), input_shape=(224,224,3)))
model7.add(Conv2D(64, kernel_size=(3, 3), input_shape=(224,224,3)))
model7.add(LeakyReLU(alpha=leaky_relu_alpha))
model7.add(MaxPooling2D(pool_size=(2, 2)))
model7.add(Dropout(0.25))

model7.add(Conv2D(64, kernel_size=(3, 3)))
model7.add(LeakyReLU(alpha=leaky_relu_alpha))
model7.add(MaxPooling2D(pool_size=(2, 2)))
model7.add(Dropout(0.25))

model7.add(Conv2D(128, kernel_size=(3, 3)))
model7.add(LeakyReLU(alpha=leaky_relu_alpha))
model7.add(MaxPooling2D(pool_size=(2, 2)))
model7.add(Dropout(0.25))

model7.add(Flatten())
model7.add(Dense(64))
model7.add(LeakyReLU(alpha=leaky_relu_alpha))
model7.add(Dropout(0.5))
model7.add(Dense(1, activation='sigmoid'))

model7.compile(loss=keras.losses.binary_crossentropy,optimizer='adam',metrics=['accuracy'])
```

```
epochs = 100
# steps_per_epoch = 10
steps_per_epoch = train_generator.n / batch_size
# steps_val = generator_val.n / batch_size
steps_test = validation_generator.n / batch_size

hist = model7.fit_generator(
    train_generator,
    steps_per_epoch=steps_per_epoch,
    epochs = epochs,
    validation_data = validation_generator,
    validation_steps=steps_test)
```
